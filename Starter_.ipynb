Web Scraping for Article Ingestion using Scrapy

```python
!pip install scrapy
```

```python
import scrapy
import json

class ArticleSpider(scrapy.Spider):
    name = 'article_spider'

    def start_requests(self):
        # List of URLs to start the crawling process
        seed_urls = [
            'https://en.wikipedia.org/wiki/Machine_learning',
            'https://arxiv.org/abs/2206.07639'
        ]
        # Create requests for each URL and initiate the parsing process
        for url in seed_urls:
            yield scrapy.Request(url=url, callback=self.parse)

    def parse(self, response):
        # Extracting the page name and creating a JSON file to store the response body
        page = response.url.split("/")[-2]
        filename = f'{page}.json'
        with open(filename, 'wb') as f:
            f.write(json.dumps(response.body))
```

```python

import spacy
from sentence_transformers import SentenceTransformer

# Load the NLP model and Sentence Transformer model
nlp = spacy.load("en_core_web_sm")
sbert_model = SentenceTransformer('all-mpnet-base-v2')

def preprocess(text):
    # Use Spacy to split the text into sentences
    doc = nlp(text)
    sentences = [sent.text for sent in doc.sents]

    # Obtain embeddings for each sentence using the SentenceTransformer model
    embeddings = sbert_model.encode(sentences)

    return {
        "sentences": sentences,
        "embeddings": embeddings
    }
```
