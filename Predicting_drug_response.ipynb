```python
# Imports
import numpy as np
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.metrics import roc_auc_score
from sklearn import datasets

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

# Generate dataset
X, y = datasets.make_classification(n_samples=1000, n_features=5, n_informative=3, n_redundant=1, n_clusters_per_class=1)

# Split data into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Build Keras sequential model
model = Sequential()
model.add(Dense(10, activation='relu', input_shape=(5,)))
model.add(Dense(1, activation='sigmoid'))

# Compile and fit model
model.compile(loss='binary_crossentropy', optimizer='adam')
model.fit(X_train, y_train, epochs=100, verbose=0)

# Evaluate model
auc = roc_auc_score(y_test, model.predict(X_test))
print(f'AUC: {auc:.3f}')

# Plot results
plt.scatter(X[:,0], y, c=y, s=40, cmap='autumn')
plt.xlabel('Feature 1')
plt.ylabel('Response')
plt.title(f'AUC: {auc:.3f}')
plt.show()
```
